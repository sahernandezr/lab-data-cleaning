#### 6. Import the posts table. 

```python
with connection:
    with connection.cursor() as cursor:
        #Read a table
        cursor.execute("SELECT * FROM posts")
        posts = cursor.fetchall()
        
posts
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
posts_df = pd.DataFrame(data=posts)
```

```python
posts_df = posts_df.rename(columns={'Id':'postId', 'OwnerUserId':'userId'})
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
users2_df = pd.DataFrame(users_df,columns=['userId', 'Reputation','Views','UpVotes','DownVotes'])
posts2_df=pd.DataFrame(posts_df,columns=['postId', 'Score','userId','ViewCount','CommentCount'])
posts2_df
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
users_posts_df=users2_df.merge(posts2_df,right_on='userId', left_on='userId')
users_posts_df
# users: 40325 rows × 5 columns
# posts: 91976 rows × 5 columns
# users_posts: 90584 rows × 9 columns
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

```python
# Count total NaN at each column in a DataFrame
print(" \nCount total NaN at each column in a DataFrame : \n\n",
      users_posts_df.isnull().sum())
```
